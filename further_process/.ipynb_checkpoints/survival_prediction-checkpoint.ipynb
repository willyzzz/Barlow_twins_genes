```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sksurv.linear_model import CoxPHSurvivalAnalysis
from sksurv.util import Surv
from sksurv.nonparametric import kaplan_meier_estimator
from sksurv.metrics import concordance_index_censored

def survival_analysis(surv, bulk, test_size=0.2, random_state=42):
    # 1. Data preprocessing: Ensure that the indices of surv and bulk are aligned to find common samples
    common_samples = surv.index.intersection(bulk.index)
    surv = surv.loc[common_samples]
    bulk = bulk.loc[common_samples]

    # Check for NaN values and remove rows with missing data
    if surv.isnull().values.any() or bulk.isnull().values.any():
        print("Warning: NaN values found in survival or gene expression data. Removing NaN rows.")
        surv.dropna(inplace=True)
        bulk.dropna(inplace=True)

    # Check for consistent survival status encoding
    print(f"Unique status values in surv: {surv['status'].unique()}")
    surv['status'] = surv['status'].astype(bool)  # Ensure status is boolean (True for death, False for survival)

    print(surv['status'])

    # 2. Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        bulk, surv, test_size=test_size, random_state=random_state, stratify=surv['status']
    )

    # Convert survival data into the format required by scikit-survival
    y_train_surv = Surv.from_dataframe("status", "time", y_train)
    y_test_surv = Surv.from_dataframe("status", "time", y_test)

    # 3. Standardize the gene expression data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Check for NaN or Inf values in scaled data
    if np.isnan(X_train_scaled).sum() > 0 or np.isinf(X_train_scaled).sum() > 0:
        raise ValueError("NaN or Inf values found in the standardized training data!")
    if np.isnan(X_test_scaled).sum() > 0 or np.isinf(X_test_scaled).sum() > 0:
        raise ValueError("NaN or Inf values found in the standardized test data!")

    # 4. Train the CoxPH model
    cox_model = CoxPHSurvivalAnalysis()
    try:
        cox_model.fit(X_train_scaled, y_train_surv)
    except ValueError as e:
        print(f"Error in fitting CoxPH model: {e}")
        return

    # Print model coefficients (risk ratio for each gene)
    coef_series = pd.Series(cox_model.coef_, index=X_train.columns)
    print("CoxPH Model Coefficients:\n", coef_series.sort_values(ascending=False))

    # Calculate C-index for the training set
    predicted_risks_train = cox_model.predict(X_train_scaled)
    time_train = y_train['time'].values
    event_train = y_train['status'].values
    c_index_train = concordance_index_censored(event_train, time_train, predicted_risks_train)
    print(f"C-Index (Training Set): {c_index_train[0]}")

    # Calculate C-index for the test set
    predicted_risks_test = cox_model.predict(X_test_scaled)
    time_test = y_test['time'].values
    event_test = y_test['status'].values
    c_index_test = concordance_index_censored(event_test, time_test, predicted_risks_test)
    print(f"C-Index (Test Set): {c_index_test[0]}")

    # Step 5: Use the model to predict and plot the survival function for a random sample from the test set
    sample = pd.DataFrame(X_test_scaled[0].reshape(1, -1), columns=X_test.columns)  # Convert the selected sample into a DataFrame
    pred_surv_func = cox_model.predict_survival_function(sample)

    # Estimate the survival probability and plot the survival function
    time_points = np.linspace(0, y_test['time'].max(), 100)
    time_points = np.clip(time_points, 0, pred_surv_func[0].domain[1])  # Ensure time_points are within valid range
    plt.step(time_points, pred_surv_func[0](time_points), where="post")
    plt.ylim(0, 1)
    plt.ylabel(r"Estimated probability of survival $\hat{S}(t)$")
    plt.xlabel("Time")
    plt.title("Survival Function for Random Sample from Test Set")
    plt.show()

    # Split training samples into high-risk and low-risk groups based on training set predictions
    risk_threshold = np.median(predicted_risks_train)  # Use the median risk score as the threshold
    high_risk_train = y_train[predicted_risks_train > risk_threshold]
    low_risk_train = y_train[predicted_risks_train <= risk_threshold]

    # Compute Kaplan-Meier survival curves for each group in the training set
    km_high_train = kaplan_meier_estimator(high_risk_train['status'], high_risk_train['time'])
    km_low_train = kaplan_meier_estimator(low_risk_train['status'], low_risk_train['time'])

    # Plot Kaplan-Meier survival curves for training set
    plt.step(km_high_train[0], km_high_train[1], where="post", label="High Risk (Train)")
    plt.step(km_low_train[0], km_low_train[1], where="post", label="Low Risk (Train)")
    plt.ylim(0, 1)
    plt.ylabel(r"Estimated probability of survival $\hat{S}(t)$")
    plt.xlabel("Time")
    plt.title("Kaplan-Meier Survival Curves (Training Set)")
    plt.legend()
    plt.show()

    # Compute Kaplan-Meier survival curves for the test set predictions
    risk_threshold_test = np.median(predicted_risks_test)  # Use the median risk score as the threshold
    high_risk_test = y_test[predicted_risks_test > risk_threshold_test]
    low_risk_test = y_test[predicted_risks_test <= risk_threshold_test]

    # Compute Kaplan-Meier survival curves for each group in the test set
    km_high_test = kaplan_meier_estimator(high_risk_test['status'], high_risk_test['time'])
    km_low_test = kaplan_meier_estimator(low_risk_test['status'], low_risk_test['time'])

    # Plot Kaplan-Meier survival curves for the test set
    plt.step(km_high_test[0], km_high_test[1], where="post", label="High Risk (Test)")
    plt.step(km_low_test[0], km_low_test[1], where="post", label="Low Risk (Test)")
    plt.ylim(0, 1)
    plt.ylabel(r"Estimated probability of survival $\hat{S}(t)$")
    plt.xlabel("Time")
    plt.title("Kaplan-Meier Survival Curves (Test Set)")
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sksurv.linear_model import CoxPHSurvivalAnalysis
from sksurv.util import Surv
from sksurv.nonparametric import kaplan_meier_estimator
from sksurv.metrics import concordance_index_censored

def survival_analysis_with_pam50(surv, pam50_labels, intervel):

    # 1. Data preprocessing: Ensure that the indices of surv and pam50_labels are aligned to find common samples
    if isinstance(pam50_labels, pd.DataFrame):
        pam50_labels = pam50_labels['PAM50']

    # Check for NaN values and drop missing data
    if surv.isnull().values.any() or pam50_labels.isnull().values.any():
        print("Warning: NaN values found in survival or PAM50 data. Removing NaN rows.")
        surv.dropna(inplace=True)
        pam50_labels.dropna(inplace=True)

    # 2. Align common samples between `surv` and `pam50_labels`
    common_samples = surv.index.intersection(pam50_labels.index)
    surv = surv.loc[common_samples]
    pam50_labels = pam50_labels.loc[common_samples]

    # 3. Convert survival data into the format required by scikit-survival
    y = Surv.from_dataframe("status", "time", surv)

    # Convert status column to boolean type
    surv['status'] = surv['status'].astype(bool)
    
    # 4. Encode the PAM50 labels into numerical values for Cox model input
    label_encoder = LabelEncoder()
    pam50_encoded = label_encoder.fit_transform(pam50_labels)

    # Create a DataFrame with the encoded PAM50 labels
    X = pd.DataFrame(pam50_encoded, columns=["PAM50"])

    # 5. Train the CoxPH model using the PAM50 subtype
    cox_model = CoxPHSurvivalAnalysis()
    try:
        cox_model.fit(X, y)
    except ValueError as e:
        print(f"Error in fitting CoxPH model: {e}")
        return

    # Print model coefficients (risk ratio for each PAM50 subtype)
    coef_series = pd.Series(cox_model.coef_, index=X.columns)
    print("CoxPH Model Coefficients:\n", coef_series.sort_values(ascending=False))

    # 6. Calculate C-index
    time = surv['time'].values
    event = surv['status'].values  # Use boolean type event status

    # Use the Cox model to predict risk scores
    predicted_risks = cox_model.predict(X)

    # Compute the C-index
    c_index = concordance_index_censored(event, time, predicted_risks)
    print(f"C-Index: {c_index[0]}")

    # 7. Plot Kaplan-Meier survival curves for each PAM50 subtype
    plt.figure(figsize=(10, 7))
    pam50_subtypes = pam50_labels.unique()  # Get unique PAM50 subtypes

    # 7. Plot Kaplan-Meier survival curves for each PAM50 subtype
    plt.figure(figsize=(10, 7))
    pam50_subtypes = pam50_labels.unique()  # Get unique PAM50 subtypes

    for subtype in pam50_subtypes:
        # Filter the samples based on the current PAM50 subtype
        mask = pam50_labels == subtype  # Create boolean mask for subtype selection
        time_subtype = surv.loc[mask, 'time']  # Select survival times for the subtype
        event_subtype = surv.loc[mask, 'status']
    
        if len(time_subtype) == 0:
            print(f"Skipping subtype {subtype} due to insufficient samples.")
            continue
    
        # Compute Kaplan-Meier estimator for the subtype
        try:
            km_time, km_prob = kaplan_meier_estimator(event_subtype, time_subtype)
        except ValueError:
            print(f"Error computing Kaplan-Meier curve for subtype: {subtype}. Skipping.")
            continue
    
        # Plot the survival curve for the subtype
        plt.step(km_time, km_prob, where="post", label=f"{subtype}")
    
    # Set labels and title
    plt.ylim(0, 1)
    plt.ylabel(r"Estimated probability of survival $\hat{S}(t)$")
    plt.xlabel("Time (months)")
    plt.title("Kaplan-Meier Survival Curves for Different PAM50 Subtypes")
    

    if intervel == "small" :
        plt.xlim(0, 120)  # Limit x-axis from 0 to 120 months
        plt.xticks(np.arange(0, 121, 20))  # Set x-ticks at intervals of 20
    else:
        pass
    
    plt.legend()
    plt.show()
```
